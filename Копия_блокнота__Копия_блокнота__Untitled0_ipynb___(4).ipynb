# Методы выявления мошеннических банковских транзакций

Проблема выявления мошеннических банковских транзакций является актуальной для финансовых организаций по всему миру. Для её решения используется широкий спектр методов, которые можно условно разделить на три группы: **правила на основе экспертного опыта**, **статистические подходы** и **методы машинного обучения**.

### 1. Правила на основе экспертного опыта

Этот подход включает в себя разработку набора жёстких правил, основанных на знаниях и опыте специалистов в области банковских операций. Например:

- Фильтрация транзакций с необычно большими суммами.
- Выявление операций, совершаемых с подозрительных IP-адресов или из географически удалённых регионов.
- Проверка частоты транзакций в короткий промежуток времени.

#### Преимущества:
- Простота реализации.
- Прозрачность и объяснимость.

#### Недостатки:
- Ограниченная гибкость: правила не адаптируются к новым видам мошенничества.
- Высокий уровень ложных срабатываний (false positives).



### 2. Статистические методы

Эти методы основаны на анализе данных о транзакциях с использованием статистических моделей. Основные подходы включают:

- **Выявление выбросов (outlier detection)**, где аномальные транзакции считаются потенциальным мошенничеством.
- **Использование вероятностных моделей** для оценки отклонений от нормального поведения клиента.

#### Преимущества:
- Возможность выявления закономерностей, невидимых для человеческого глаза.
- Широкое применение в анализе данных.

#### Недостатки:
- Требуется ручная настройка порогов для классификации.
- Сложности в обработке больших объёмов данных.



### 3. Методы машинного обучения

Машинное обучение является одним из наиболее перспективных подходов к решению задачи. Оно позволяет моделям адаптироваться к данным и выявлять скрытые закономерности. Применяемые алгоритмы включают:

- **Модели классификации**: логистическая регрессия, решающие деревья, случайный лес (Random Forest), градиентный бустинг (XGBoost, LightGBM, CatBoost).
- **Нейронные сети**: глубокие архитектуры (Deep Learning), такие как рекуррентные нейронные сети (RNN) и сверточные нейронные сети (CNN) для обработки временных и пространственных данных.
- **Ансамблевые методы**: комбинация нескольких моделей для повышения точности.

#### Преимущества:
- Высокая точность и способность выявлять сложные зависимости.
- Возможность автоматической обработки больших массивов данных.

#### Недостатки:
- Требуются значительные вычислительные ресурсы.
- Необходимость в качественной подготовке данных и настройке модели.



### 4. Гибридные подходы

На практике часто применяются гибридные методы, которые объединяют преимущества различных подходов. Например, правила на основе экспертного опыта могут быть использованы для предварительного фильтрации данных, после чего на очищенных данных применяется алгоритм машинного обучения.



#### Сравнительный анализ

Сравнение существующих методов показывает, что наибольший потенциал имеют подходы на основе машинного обучения благодаря их способности автоматически адаптироваться к изменениям в данных и обнаруживать новые формы мошенничества. Однако их эффективность напрямую зависит от качества данных и использования адекватных методов предобработки и настройки.

##**1.3 Обоснование возможности использования методов машинного обучения для решения задачи**

Это делает традиционные подходы, основанные на фиксированных правилах или статистических методах, менее эффективными. Рассмотрим основные аспекты, обосновывающие возможность применения МЛ для решения данной задачи.

### 1. Обработка больших объемов данных

Современные банки ежедневно обрабатывают миллионы транзакций, каждая из которых характеризуется множеством признаков (сумма, время, географическое местоположение, тип операции и т. д.). Методы МЛ способны эффективно работать с такими большими объемами данных, выявляя скрытые зависимости, которые неочевидны для традиционных методов.

### 2. Обнаружение сложных паттернов

МЛ позволяет обнаруживать сложные и нелинейные зависимости между признаками. Например:

- Аномальное поведение пользователя (необычная геолокация, нетипичная сумма).

## 3. Обучение на исторических данных

МЛ-алгоритмы обучаются на исторических данных, содержащих информацию о нормальных и мошеннических транзакциях. Это позволяет моделям "запомнить" типичные сценарии мошенничества и применять эти знания к новым данным. Используемый набор данных «Банковские транзакции.csv» предоставляет возможность построить качественную модель за счёт меток классов (мошенничество/немошенничество).

### 4. Адаптация к изменениям

Мошенники постоянно разрабатывают новые схемы обхода защитных систем. Методы МЛ, особенно онлайн-обучение и методы глубокого обучения, способны динамически адаптироваться к новым данным, обновляя свои модели на основе поступающих транзакций.

### 5. Снижение уровня ложных срабатываний

Одной из ключевых задач является минимизация ложных срабатываний, чтобы не блокировать легитимные транзакции. МЛ-алгоритмы способны учитывать большое количество факторов и использовать контекст, что позволяет снизить уровень ошибок по сравнению с жесткими правилами.

### 6. Автоматизация процесса выявления мошенничества

МЛ позволяет автоматизировать процесс выявления аномалий, что снижает необходимость ручного вмешательства и повышает эффективность работы системы. Это особенно важно для обработки большого потока данных в режиме реального времени.

Современные библиотеки и фреймворки (например, Scikit-learn, TensorFlow, PyTorch) делают внедрение МЛ-решений доступным. Они предоставляют готовые реализации популярных алгоритмов, что позволяет сосредоточиться на настройке и адаптации модели под конкретные задачи.

## Выбор метода обучения: случайный лес (Random Forest)

В рамках данной курсовой работы для выявления мошеннических банковских транзакций выбран метод обучения **случайный лес (Random Forest)**. Этот алгоритм является одним из наиболее эффективных методов машинного обучения для задач классификации, особенно в условиях, когда данные имеют сложную структуру и множество признаков.

## Принцип работы метода случайный лес

### 1. Генерация подвыборок данных
- Алгоритм использует метод бутстрепинга для создания случайных подвыборок данных из исходного набора.
- Каждая подвыборка используется для обучения одного решающего дерева.

### 2. Построение решающих деревьев
- Для каждой подвыборки строится независимое дерево.
- При создании каждого узла дерева используется случайное подмножество признаков для выбора наилучшего разделения данных.
- Этот процесс снижает корреляцию между деревьями, увеличивая обобщающую способность модели.

### 3. Комбинирование результатов
- Для задачи классификации применяется метод голосования большинства: итоговый класс определяется на основе большинства предсказаний отдельных деревьев.
- В задачах регрессии используется усреднение предсказаний деревьев.

### 4. Устойчивость к переобучению
- Благодаря ансамблированию и случайности, алгоритм сглаживает ошибки отдельных деревьев и минимизирует риск переобучения.

## Основные характеристики алгоритма

### 1. Построение деревьев
- Алгоритм создаёт множество независимых решающих деревьев.
- Каждое дерево обучается на случайной подвыборке данных, полученной методом бутстрепинга.

### 2. Выбор признаков
- Для каждого узла дерева используется случайное подмножество признаков, что снижает корреляцию между деревьями и улучшает обобщающую способность модели.

### 3. Голосование
- Для классификации применяется метод "голосования большинства": класс, который предсказали большинство деревьев, становится итоговым.

### 4. Устойчивость к переобучению
- Случайный лес сглаживает ошибки отдельных деревьев, что делает модель устойчивой к переобучению.

## Преимущества случайного леса

### 1. Высокая точность
- За счёт ансамблирования случайный лес обладает высокой точностью и хорошо справляется с нелинейными зависимостями.

### 2. Устойчивость к шуму
- Использование случайных подвыборок данных и признаков делает алгоритм менее чувствительным к выбросам и шуму в данных.

### 3. Интерпретируемость
- Алгоритм предоставляет оценки важности признаков, что позволяет понять, какие факторы больше всего влияют на результат.

### 4. Гибкость
- Работает как с числовыми, так и с категориальными данными.

### 5. Параллельная обработка
- Деревья можно обучать независимо, что ускоряет процесс обучения на многопроцессорных системах.

## **2.3 Описание и математическое выражение функции потерь для обучения выбранной модели машинного обучения**

## Функция потерь для модели случайного леса

Модель случайного леса (Random Forest) состоит из множества решающих деревьев, каждое из которых минимизирует свою функцию потерь. Основная идея случайного леса заключается в ансамблировании нескольких деревьев для улучшения общей производительности модели. Для случайного леса общая функция потерь определяется на уровне отдельных деревьев, а итоговое предсказание выбирается по **голосованию большинства**.


## Что такое функция потерь?

**Функция потерь** — это математическое выражение, которое измеряет ошибку модели машинного обучения, то есть разницу между предсказанным моделью значением и истинным значением. Она определяет, насколько хорошо или плохо модель справляется с поставленной задачей.



### 1. Описание функции потерь для случайного леса

Для обучения каждого дерева в случайном лесу используется функция потерь, такая же, как в решающих деревьях. На этапе разбиения узлов дерева применяется **критерий информативности**, например:
- **Индекс Джини** (Gini Impurity),
- **Энтропия** (Information Gain).

Эти критерии используются для минимизации неоднородности классов в узлах дерева.


### 2. Индекс Джини

**Индекс Джини** измеряет вероятность того, что выбранный случайным образом объект будет неправильно классифицирован, если его классифицировать случайно по распределению классов.

Формула для индекса Джини:
$$
G = 1 - \sum_{k=1}^{K} p_k^2
$$
где:
- $p_k$ — вероятность того, что объект относится к классу \(k\) (доля объектов класса \(k\) в текущем узле),
- K — количество классов.

**Минимизация индекса Джини** приводит к выбору разбиения, которое делает узлы как можно более "чистыми".



### 3. Энтропия

Энтропия измеряет степень неопределённости (или хаоса) в данных. Чем выше энтропия, тем менее однороден узел.

Формула для энтропии:
$$
H = - \sum_{k=1}^{K} p_k \log_2(p_k)
$$


где:
 - $p_k$ — вероятность того, что объект принадлежит классу k.

Энтропия уменьшается, если узел становится более однородным, а целью модели является минимизация энтропии.



### 4. Общая функция потерь случайного леса

Функция потерь случайного леса является **ансамблированием потерь отдельных деревьев**. Итоговая функция потерь минимизируется за счёт:
1. Выбора случайных подвыборок данных для обучения каждого дерева,
2. Итеративного разбиения узлов каждого дерева на основе критерия информативности (индекс Джини или энтропия).

Если $L(Ti)$ — функция потерь \(i\)-го дерева в лесу, то общая функция потерь случайного леса:
$$
L_{\text{RF}} = \frac{1}{N} \sum_{i=1}^{N} L(T_i)
$$
где:
- \(N\) — общее количество деревьев.



## Функция потерь в задаче классификации

Для задачи классификации основная цель функции потерь — минимизировать ошибку предсказания класса. Это достигается путём:

### 1. Оптимизации разбиения узлов дерева:
- С помощью индекса Джини или энтропии на каждом шаге разбиения минимизируется неоднородность узлов,
- Узел считается "чистым", если все объекты в нём принадлежат к одному классу (\(G = 0\) или \(H = 0\)).

### 2. Использования ансамбля деревьев:
- Общая ошибка модели уменьшается за счёт усреднения предсказаний нескольких деревьев или голосования большинства.

В задаче классификации результатом случайного леса является вероятность принадлежности объекта к каждому классу:
$$
\hat{y}_i = \text{mode}(T_1(x_i), T_2(x_i), \dots, T_k(x_i))
$$
где:
- $(T_1, T_2, \dots, T_k)$ — решающие деревья,
- $\text{mode}$ — наиболее часто встречающееся предсказание.


## Итог

Функция потерь случайного леса:
1. Минимизирует неоднородность узлов с помощью индекса Джини или энтропии,
2. Для задачи классификации направлена на минимизацию ошибок в определении класса,
3. На уровне ансамбля минимизирует совокупную ошибку путём усреднения или голосования большинства, что делает модель устойчивой к шуму и переобучению.


##2.4 Описание метода обучения выбранной модели машинного обучения.


### Метод обучения случайного леса (Random Forest)

Случайный лес (Random Forest) — это мощный и гибкий алгоритм машинного обучения, который решает задачи классификации и регрессии. Этот метод основан на ансамблировании (объединении) большого числа решающих деревьев. Его ключевая идея заключается в том, что комбинация простых моделей (деревьев) даёт более точные и устойчивые результаты.



### Как устроен случайный лес

1. **Ансамбль деревьев**:  
   Случайный лес состоит из множества решающих деревьев, которые строятся независимо друг от друга. Эти деревья объединяются для формирования окончательного предсказания.

2. **Рандомизация**:  
   Чтобы деревья в ансамбле были разнообразными, случайный лес использует два основных источника случайности:
   - **Случайная подвыборка данных**: каждое дерево обучается на уникальной подвыборке данных, созданной методом бутстрепинга (выбор с возвращением).
   - **Случайный выбор признаков**: для каждого разбиения узла дерева случайный лес выбирает случайное подмножество признаков, что снижает корреляцию между деревьями.

3. **Голосование или усреднение**:  
   - Для задач классификации используется **голосование большинства**: класс, предсказанный большинством деревьев, становится итоговым результатом.
   - Для задач регрессии используется **усреднение** предсказаний всех деревьев.

### Как работает обучение случайного леса

#### Этапы обучения:

1. **Подготовка данных**:  
   Данные разделяются на обучающую и тестовую выборки. Обучающая выборка используется для построения деревьев, а тестовая — для оценки качества модели.

2. **Бутстрепинг**:  
   Для каждого дерева случайным образом выбирается подвыборка данных из обучающей выборки. Некоторые объекты могут встречаться несколько раз, а часть объектов может не войти в подвыборку (Out-of-Bag).

3. **Построение дерева**:  
   Каждое дерево строится на своей подвыборке данных, используя случайное подмножество признаков. Алгоритм дерева включает:
   - **Разбиение узлов**: на каждом шаге выбирается признак и порог, которые максимизируют критерий информативности (например, минимизируют индекс Джини или энтропию).
   - **Остановка роста**: дерево растёт до заданной глубины или пока в узле не останется минимальное количество объектов.

4. **Формирование ансамбля**:  
   После обучения всех деревьев их предсказания объединяются:
   - В задачах классификации: итоговый класс определяется большинством голосов.
   - В задачах регрессии: итоговое значение — это среднее предсказаний всех деревьев.

  ### Индекс Джини

  Индекс Джини измеряет вероятность того, что выбранный случайным образом объект будет неправильно классифицирован, если его классифицировать случайно по распределению классов.

  Формула для индекса Джини:

  $$

  - $p_k$ — вероятность того, что объект относится к классу \(k\) (доля объектов класса \(k\) в текущем узле).
  - K — количество классов.

  Минимизация индекса Джини приводит к выбору разбиения, которое делает узлы как можно более "чистыми".

     
   - **Энтропия (Information Gain)**: измеряет хаос в распределении классов.
     
     $$H = - \sum_{k=1}^{K} p_k \log_2(p_k)$$

2. **Гиперпараметры случайного леса**:

   - `max_depth`: максимальная глубина дерева.  
     - Задаёт максимальное количество уровней в деревьях. Ограничение глубины предотвращает переобучение, особенно на шумных данных.
   - `max_features`: количество признаков, рассматриваемых при разбиении узла.  
     - Это случайно выбранное подмножество признаков, используемое для определения лучшего разбиения. Возможные значения:
       - `"sqrt"` (по умолчанию): квадратный корень из общего числа признаков.
       - `"log2"`: логарифм по основанию 2 от числа признаков.
       - Целое число или дробное значение (например, `0.5` означает 50% признаков).
   - `min_samples_split`: минимальное количество объектов для разбиения узла.  
     - Если в узле содержится меньше объектов, чем указано в этом параметре, разбиение не производится. Используется для предотвращения переобучения.
   - `min_samples_leaf`: минимальное количество объектов в листовом узле.  
     - Определяет минимальный размер конечных узлов дерева. Позволяет уменьшить глубину дерева и избежать разбиения узлов с малым количеством объектов.
   - `bootstrap`: использовать ли бутстреп для создания подвыборок данных.  
     - Если `True`, то каждое дерево обучается на случайной подвыборке данных (с возвращением).
   - `random_state`: начальное значение для генератора случайных чисел.  
     - Позволяет воспроизводить результаты.
   - `class_weight`: веса классов.  
     - Используется для компенсации дисбаланса классов. Возможные значения:
       - `"balanced"`: веса автоматически рассчитываются, чтобы классы были равномерно представлены.
       - Пользовательский словарь, определяющий веса для каждого класса.

    - `base_estimator_`: базовый объект, используемый для построения деревьев (по умолчанию `DecisionTreeClassifier`).

**Выбросы** (outliers) — это наблюдения в наборе данных, которые значительно отличаются от остальных значений и находятся вне типичного диапазона. Такие данные могут быть результатом ошибок измерения, некорректного ввода или отражать реальные, но редкие явления.


### Как определить выбросы?

1. **Статистические методы:**
   - **Метод 3σ:** Выбросы находятся за пределами трёх стандартных отклонений от среднего значения.
   - **Интерквартильный размах (IQR):** Выбросы определяются как значения, выходящие за пределы диапазона:
  Выбросы определяются как значения, выходящие за пределы диапазона:

  $$
  [Q_1 - 1.5 \cdot IQR, \; Q_3 + 1.5 \cdot IQR]
  $$

  где:

  - $Q_1$ — первый квартиль (25-й перцентиль),
  - $Q_3$ — третий квартиль (75-й перцентиль),
  - $(IQR = Q_3 - Q_1)$ — интерквартильный размах.

2. **Графические методы:**
   - **Диаграммы размаха (boxplot):** Выбросы визуализируются как точки за пределами "усов" диаграммы.
   - **Scatter Plot:** Аномалии могут быть видны как точки, значительно удалённые от основной группы данных.

3. **Машинное обучение:**
   - Использование алгоритмов обнаружения выбросов, таких как изоляционные леса (Isolation Forest), кластеризация (например, DBSCAN) или метод локальной факторной аномалии (LOF).

```python
import matplotlib.pyplot as plt

# Построение боксплотов для всех признаков на одном полотне и удаление выбросов
plt.figure(figsize=(10, 50))  # Устанавливаем размер графика для визуализации

# Проходим по каждому признаку из списка числовых признаков
for i, column in enumerate(numeric_columns, 1):  # enumerate даёт индекс и название признака
    # Расчет первого (Q1) и третьего (Q3) квартилей
    Q1 = df[column].quantile(0.25)  # Первый квартиль (25-й процентиль)
    Q3 = df[column].quantile(0.75)  # Третий квартиль (75-й процентиль)
    IQR = Q3 - Q1  # Интерквартильный размах (разница между Q3 и Q1)

    # Определение границ выбросов
    lower_bound = Q1 - 1.5 * IQR  # Нижняя граница (в пределах 1.5 интерквартильных размахов от Q1)
    upper_bound = Q3 + 1.5 * IQR  # Верхняя граница (в пределах 1.5 интерквартильных размахов от Q3)

    # Подсчёт количества выбросов в текущем признаке
    outliers = ((df[column] < lower_bound) | (df[column] > upper_bound)).sum()

    # Удаление строк с выбросами для текущего признака
    df = df[~((df[column] < lower_bound) | (df[column] > upper_bound))]

    # Построение боксплота для текущего признака
    plt.subplot(len(numeric_columns), 1, i)  # Создаем отдельный график для каждого признака
    plt.boxplot(df[column], vert=False, patch_artist=True)  # Рисуем горизонтальный боксплот
    plt.title(f'{column} (Outliers: {outliers})')  # Указываем имя признака и количество выбросов
    plt.xlabel('Value')  # Метка оси X
    plt.grid(axis='x')  # Добавляем сетку для удобства чтения графика

# Уплотнение макета графиков
plt.tight_layout()

# Отображение всех графиков
plt.show()
```

Квартиль — это статистический показатель, который делит упорядоченный набор данных на четыре равные части. Каждая часть содержит примерно 25% данных.

Первый квартиль (Q1):

* Значение, ниже которого находится 25% всех данных.
* Это 25-й процентиль.

Медиана (Q2):

* Это второй квартиль, делящий данные пополам.
* Значение, ниже которого находится 50% всех данных.
* Медиана также известна как 50-й процентиль.

Третий квартиль (Q3):

* Значение, ниже которого находится 75% всех данных.
* Это 75-й процентиль.

##Балансировка классов


**Балансировка классов** — это процесс корректировки распределения данных в задаче классификации, где количество наблюдений в разных классах сильно отличается. Эта ситуация известна как **дисбаланс классов**, и она может негативно повлиять на производительность модели машинного обучения.

## Почему дисбаланс классов — это проблема?

1. **Смещение модели:**
   - Модель склонна предсказывать преобладающий класс, так как он чаще встречается.
   - Например, если 95% транзакций нормальные и только 5% мошеннические, модель может "научиться" всегда предсказывать класс "нормальные" и достигать 95% точности, игнорируя редкий класс.

2. **Снижение метрик качества:**
   - Метрика **точность (accuracy)** становится менее информативной, так как она не учитывает вклад меньшинства.
   - Важно использовать метрики, такие как **F1-мера**, **Precision**, **Recall** или **ROC-AUC**, чтобы лучше оценивать модель в условиях дисбаланса.


## Методы балансировки классов

### 1. Переподбор данных (Resampling)

#### 1.1. **Oversampling (увеличение данных для меньшинства):**
- Искусственное увеличение количества наблюдений для меньшинства.
- **Примеры:**
  - Копирование существующих данных редкого класса.
  - Использование методов, таких как **SMOTE (Synthetic Minority Oversampling Technique)**, для генерации синтетических данных.

#### 1.2. **Undersampling (уменьшение данных для большинства):**
- Удаление части данных из преобладающего класса, чтобы выровнять их количество с редким классом.
- **Недостаток:** Возможная потеря полезной информации.

#### 1.3. **Комбинированный метод:**
- Совмещение oversampling для меньшинства и undersampling для большинства.


### 2. Взвешивание классов (Class Weighting)
- Назначение большего веса редкому классу при обучении модели.
- Это заставляет алгоритм уделять больше внимания редкому классу.
- **Пример:** В Scikit-learn можно задать параметр `class_weight='balanced'`.

### 3. Генерация синтетических данных
- Использование продвинутых методов, таких как:
  - **SMOTE (Synthetic Minority Oversampling Technique):** Генерация синтетических образцов на основе интерполяции между ближайшими соседями редкого класса.
  - **ADASYN (Adaptive Synthetic Sampling):** Расширение SMOTE с учётом сложности распределения данных.

##Коррелирующие признаки

**Коррелирующие признаки** — это признаки (переменные) в наборе данных, которые имеют сильную статистическую зависимость друг от друга. Это означает, что изменение значения одного признака связано с изменением значения другого. В задачах машинного обучения корреляция между признаками может стать проблемой, так как она приводит к избыточности информации, увеличивает сложность модели и может негативно сказаться на её производительности.

## Проблемы, связанные с коррелирующими признаками

1. **Избыточность информации:**
   - Коррелирующие признаки не добавляют новой информации, так как несут схожие данные.

2. **Увеличение сложности модели:**
   - Избыточные признаки увеличивают размерность данных, что может привести к "проклятию размерности" и усложнению обучения модели.

3. **Переобучение:**
   - Модель может сосредоточиться на коррелирующих признаках, считая их важными, что снижает её обобщающую способность.

4. **Нестабильность коэффициентов:**
   - В линейных моделях (например, линейной регрессии) коррелирующие признаки могут вызывать мультиколлинеарность, что делает модель нестабильной и трудной для интерпретации.



## Как обработать коррелирующие признаки?

1. **Удаление избыточных признаков:**
   - Выбор одного наиболее информативного признака из группы коррелирующих.

2. **Снижение размерности:**
   - Применение методов, таких как PCA (метод главных компонент), для преобразования коррелирующих признаков в новые, некоррелированные переменные.

3. **Регуляризация:**
   - Использование регуляризационных методов (например, Lasso или Ridge регрессии), чтобы уменьшить влияние коррелирующих признаков.

```python
import pandas as pd  # Импорт библиотеки для работы с данными

def drop_features_by_correlation(df, target_column, min_corr=0.1, max_corr=0.9):
    """
    Удаляет признаки из датафрейма, если их корреляция с целевой переменной меньше min_corr
    или больше max_corr.

    :param df: pd.DataFrame, входной датафрейм
    :param target_column: str, название целевого столбца
    :param min_corr: float, минимальное значение корреляции (по модулю)
    :param max_corr: float, максимальное значение корреляции (по модулю)
    :return: pd.DataFrame, датафрейм с удалёнными признаками
    """

    # Вычисляем корреляционную матрицу для всех признаков
    correlation_matrix = df.corr()

    # Извлекаем корреляцию признаков с целевой переменной
    target_correlation = correlation_matrix[target_column].abs()

    # Отбираем признаки, которые имеют корреляцию меньше min_corr или больше max_corr
    to_drop = target_correlation[
        (target_correlation < min_corr) | (target_correlation > max_corr)
    ].index  # Получаем имена таких признаков

    to_drop = to_drop.drop(target_column)  # Исключаем целевой столбец из списка для удаления

    # Печатаем список признаков, которые будут удалены
    print(f"Признаки для удаления: {list(to_drop)}")

    # Удаляем признаки с неподходящей корреляцией
    df = df.drop(columns=to_drop)

    # Возвращаем очищенный датафрейм
    return df
```

Отбираются признаки, корреляция которых:
* Меньше минимального значения (min_corr) — они мало связаны с целевым классом и могут не быть полезными для модели.
* Больше максимального значения (max_corr) — слишком высокая корреляция может привести к дублированию информации.


Зачем это нужно:
* Удаление нерелевантных признаков:

  1. Признаки с низкой корреляцией (например, <0.1) не содержат значимой информации для предсказания целевого класса. Их удаление улучшает производительность модели.

* Устранение мультиколлинеарности:

  1. Признаки с высокой корреляцией между собой дублируют информацию. Удаление
таких признаков снижает избыточность данных.

* Упрощение модели:

  1. Удаление ненужных признаков уменьшает размерность данных, что снижает риск переобучения.

## Формула

Коэффициент корреляции Пирсона вычисляется по формуле:

$$
r = \frac{\text{Cov}(X, Y)}{\sigma_X \cdot \sigma_Y}
$$

Где:
- $Cov(X, Y)$ — ковариация двух признаков $ X $ и $ Y $
- $sigma_X, sigma_Y$ — стандартные отклонения признаков $ X $ и $ Y $ .


## Шаги вычисления

### 1. Средние значения
Сначала вычисляем средние значения для каждого признака:

$$
\bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i, \quad \bar{Y} = \frac{1}{n} \sum_{i=1}^{n} Y_i
$$

Где:
- $n$ — количество наблюдений,
- $X_i, Y_i$ — значения признаков X и Y для i-го наблюдения.



### 2. Ковариация

Ковариация показывает, как два признака изменяются вместе:

$$
\text{Cov}(X, Y) = \frac{\sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})}{n - 1}
$$



### 3. Стандартное отклонение

Стандартное отклонение показывает, насколько значения признаков отклоняются от их среднего:

$$
\sigma_X = \sqrt{\frac{\sum_{i=1}^{n} (X_i - \bar{X})^2}{n - 1}}, \quad \sigma_Y = \sqrt{\frac{\sum_{i=1}^{n} (Y_i - \bar{Y})^2}{n - 1}}
$$



### 4. Итоговая формула

После вычисления ковариации и стандартных отклонений подставляем значения в формулу корреляции:

$$
r = \frac{\text{Cov}(X, Y)}{\sigma_X \cdot \sigma_Y}
$$


**Среднее значение** показывает центральное (среднее) положение данных. Это сумма всех значений признака, делённая на их количество.

### Формула:
$$
\bar{X} = \frac{\sum_{i=1}^{n} X_i}{n}
$$

### Где:
- $X_i$ — значения  признака $X$,
- $n — количество наблюдений (строк).$

### Пример:
Допустим, у нас есть данные: \(X = [2, 4, 6]\).  
Среднее значение:
$$
\bar{X} = \frac{2 + 4 + 6}{3} = 4
$$

## 2. Ковариация

**Ковариация** измеряет, как два признака изменяются вместе. Она показывает, связаны ли они положительно (оба увеличиваются) или отрицательно (один увеличивается, другой уменьшается).

### Формула:
$$
\text{Cov}(X, Y) = \frac{\sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})}{n - 1}
$$

### Где:
- $X, Y$ — два признака,
- $bar{X}$, $bar{Y}$ — их средние значения.

### Результат:
- **Положительная ковариация:** Если \(X\) увеличивается, \(Y\) тоже увеличивается.
- **Отрицательная ковариация:** Если \(X\) увеличивается, \(Y\) уменьшается.
- **Нулевая ковариация:** Между признаками нет связи.

### Пример:
Данные: \(X = [1, 2, 3]\), \(Y = [2, 4, 6]\).  
Средние значения: $bar{X} = 2$, $bar{Y} = 4$.

Считаем ковариацию:
$$
\text{Cov}(X, Y) = \frac{(1-2)(2-4) + (2-2)(4-4) + (3-2)(6-4)}{3 - 1} = \frac{2 + 0 + 2}{2} = 2
$$

Положительная ковариация $2$ показывает, что \(X\) и \(Y\) увеличиваются вместе.


## 3. Стандартное отклонение

**Стандартное отклонение** измеряет, насколько значения признака разбросаны относительно среднего. Чем больше стандартное отклонение, тем сильнее разброс.

### Формула:
$$
\sigma_X = \sqrt{\frac{\sum_{i=1}^{n} (X_i - \bar{X})^2}{n - 1}}
$$

### Где:

- $X_i$ — значения признака,
- $bar{X}$ — среднее значение,
- $n$ — количество наблюдений.

### Результат:
- **Малое стандартное отклонение:** Значения близки к среднему.
- **Большое стандартное отклонение:** Значения сильно разбросаны.

### Пример:
Данные: \(X = [1, 2, 3]\).  
Среднее значение: \(\bar{X} = 2\).

Считаем стандартное отклонение:
$$
sigma_X = \sqrt{\frac{(1-2)^2 + (2-2)^2 + (3-2)^2}{3 - 1}} = \sqrt{\frac{1 + 0 + 1}{2}} = \sqrt{1} = 1
$$

Стандартное отклонение равно $1$, что показывает небольшой разброс значений.

##**3.2 Формирование обучающей и тестовой выборок и обучение модели с гиперпараметрами по умолчанию**

**Гиперпараметры** — это параметры модели машинного обучения, которые не обучаются на данных, а задаются пользователем **до начала обучения**. Они определяют, как именно модель будет обучаться, и могут сильно влиять на её производительность.

Используем параметры по умолчанию для первичной оценки модели.

##Разделим данные на тестовые и обучающие

1. **Обучающая выборка (Training Set)** — данные, которые используются для обучения модели.
2. **Тестовая выборка (Test Set)** — данные, которые используются для оценки качества обученной модели.

3. **Целевая переменная (Target Variable)** — это переменная, которую модель машинного обучения пытается предсказать. Она указывает на тот результат или категорию, к которому относится каждая запись в данных. Целевая переменная обозначает **выходные данные** в задаче машинного обучения.


Разделяем выборку при помощи **train_test_split**

1. X и y:

  * X — признаки (входные данные, матрица или DataFrame),
  * y — целевая переменная (вектор или Series).

2. test_size:

  * Доля данных, которая попадёт в тестовую выборку.
  Значение по умолчанию: test_size=None (но при этом train_size=0.75, то есть 25% данных уйдёт на тест).

 *  **Пример:** test_size=0.3 означает, что 30% данных пойдёт на тест, а 70% — на обучение.

3. train_size:

  * Доля данных, которая попадёт в обучающую выборку.
  Если задано, должно быть совместимо с test_size.

4. random_state:

  * Число, определяющее случайность разбиения.
  Используется для воспроизводимости результатов (одно и то же значение random_state даст одинаковое разбиение при каждом запуске).

##Возвращаемые значения

2. X_test: Признаки для тестирования.
3. y_train: Целевая переменная для обучения.
4. y_test: Целевая переменная для тестирования.


**`RandomForestClassifier`** — это реализация алгоритма **случайного леса (Random Forest)** из библиотеки **scikit-learn**. Этот алгоритм используется для задач классификации, где цель — предсказать класс объекта (например, мошенническая или нормальная транзакция).



## Как работает `RandomForestClassifier`?

- Случайный лес строит множество независимых решающих деревьев.
- Каждое дерево обучается на случайной подвыборке данных.
- Итоговый результат предсказания определяется методом голосования большинства среди деревьев.


### 2. **Процесс работы**

#### Этап 1: Бутстреппинг (Bootstrap Sampling)
- Для каждого дерева случайно выбирается подвыборка данных из обучающего набора с возвратом.
- Эта подвыборка может содержать повторяющиеся записи.

#### Этап 2: Построение деревьев
- В каждом узле дерева для разбиения выбирается случайное подмножество признаков.
- Узел делится по признаку и порогу, которые минимизируют неоднородность классов (например, с использованием индекса Джини или энтропии).

#### Этап 3: Голосование
- Для классификации каждое дерево голосует за свой предсказанный класс.
- Итоговое предсказание — класс, который получил наибольшее количество голосов (метод большинства).


## Основные гиперпараметры

1. **`n_estimators` (по умолчанию `100`):**
   - Количество деревьев в лесу.
   - Увеличение числа деревьев повышает точность, но замедляет обучение.

2. **`max_depth` (по умолчанию `None`):**
   - Максимальная глубина дерева.
   - Если не указано, дерево строится до полного разбиения узлов.

3. **`max_features` (по умолчанию `'sqrt'`):**
   - Количество признаков, которое анализируется для разбиения каждого узла.

4. **`random_state`:**
   - Фиксирует случайность для воспроизводимости результатов.

5. **`n_jobs` (по умолчанию `None`):**
   - Количество процессоров, используемых для параллельного обучения деревьев.
   - Значение `-1` использует все доступные процессоры.




```python
from sklearn.model_selection import train_test_split  # Импорт функции для разделения данных на обучающую и тестовую выборки
from sklearn.ensemble import RandomForestClassifier  # Импорт классификатора случайного леса
from sklearn.metrics import classification_report, confusion_matrix  # Метрики для оценки качества модели
import numpy as np  # Импорт библиотеки numpy для работы с массивами (здесь пока не используется)

# Разделяем признаки (X) и целевую переменную (y)
X = df.drop(columns=['Class'])  # Удаляем столбец 'Class', оставляя только признаки
y = df['Class']  # Целевая переменная (мошеннические/нормальные транзакции)

# Разделяем данные на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(
    X, y,  # Входные данные (X) и целевая переменная (y)
    test_size=0.3,  # Размер тестовой выборки — 30% от общего объёма данных
    random_state=21  # Фиксируем случайное состояние для воспроизводимости
)

# Создаём модель случайного леса
rf_model = RandomForestClassifier(
    oob_score=True,  # Используем "Out-of-Bag" оценку для проверки качества модели
    n_jobs=-1,  # Используем все доступные процессоры для ускорения обучения
    random_state=21  # Фиксируем случайное состояние для воспроизводимости
)
X_train_main = X_train.copy()
X_test_main =  X_test.copy()
y_train_main = y_train.copy()
y_test_main = y_test.copy()

# Обучаем модель на обучающих данных
rf_model.fit(X_train, y_train)  # Передаём обучающие признаки и целевую переменную
```

##Кросс-валидация


Кросс-валидация — это метод оценки качества модели, который проверяет её устойчивость и способность обобщать данные. Она позволяет:

1. Проверить стабильность модели: Оценивается, насколько результаты на разных разбиениях данных близки друг к другу. Низкая дисперсия указывает на стабильность.

2. Избежать переобучения: Тестирование на новых подвыборках помогает убедиться, что модель не "запоминает" обучающие данные, а способна предсказывать на новых данных.

3. Получить более точную оценку: Среднее значение кросс-валидации предоставляет надёжное представление о производительности модели, в отличие от однократного тестирования.

4. Обеспечить равномерное распределение данных: Разбиение данных на части позволяет учесть их дисбаланс или особенности, повышая надёжность оценки.

5. Подтвердить качество модели: Сравнение метрик кросс-валидации и тестовой выборки подтверждает корректное обучение модели и надёжность результатов.

Кросс-валидация позволяет:
1. **Избежать переобучения:**
   - Кросс-валидация позволяет проверить, как модель будет работать на данных, которые она "не видела" при обучении.

2. **Оценить производительность:**
   - Вместо одного случайного разделения на обучающую и тестовую выборки, кросс-валидация проверяет модель на нескольких разных разбиениях данных.


### Этапы работы:

#### 1. Разделение данных
- Данные делятся на **\(k\) равных частей (фолдов)**.
- Обычно \(k = 5\) или \(k = 10\).

#### 2. Итеративное обучение и тестирование
- На каждой итерации:
  - Одна часть данных (1 фолд) используется как **тестовая выборка**.
  - Остальные \(k-1\) фолдов используются как **обучающая выборка**.
- Модель обучается на обучающей выборке и оценивается на тестовой.

#### 3. Усреднение результата
- Считаются метрики (например, точность, F1-мера) для каждой итерации.
- Итоговая оценка модели — это среднее значение метрик по всем фолдам.


Итоговая метрика:

$$
\text{Средняя метрика} = \frac{\text{Метрика 1 + Метрика 2 + ... + Метрика k}}{k}
$$

где $k$ - количество фолдов

```python
from sklearn.model_selection import cross_val_score  # Импорт функции для проведения кросс-валидации

# Используем 5-кратную кросс-валидацию
cv_scores = cross_val_score(  # Функция возвращает оценки качества модели для каждой из 5 частей
    rf_model,  # Модель случайного леса, которую мы хотим проверить
    X,         # Все признаки из датасета
    y,         # Целевая переменная
    cv=5       # Число разбиений данных (5-кратная кросс-валидация)
)

# Выводим оценки для каждой итерации кросс-валидации
print(f"Кросс-валидация: {cv_scores}")

# Вычисляем среднюю оценку по всем итерациям кросс-валидации
print(f"Средняя оценка кросс-валидации: {cv_scores.mean()}")
```

1. Модель стабильна: Результаты кросс-валидации демонстрируют, что модель уверенно предсказывает вне зависимости от разбиения данных.
2. Высокое качество: Средняя точность 99.89% говорит о том, что модель эффективно справляется с задачей классификации.
3. Отсутствие переобучения (предположение): Стабильные результаты на кросс-валидации позволяют предположить, что модель не переобучается.

##OOB-оценка качества модели



**Out-of-Bag (OOB) оценка** — это встроенный механизм в модели ансамблей , который позволяет оценить качество модели без использования дополнительной тестовой выборки.

**Бутстреппинг** — это статистический метод, который используется для создания подвыборок из исходных данных с возвратом. Он применяется для моделирования вариативности данных и оценки качества модели.


1. Данные делятся методом бутстреппинга
  * Для каждого дерева случайного леса берётся случайная подвыборка данных из исходного датасета.
  Остальные данные из датасета, которые не попали в подвыборку, становятся OOB-записями для этого дерева

2. Обучение на подвыборке
  * Дерево обучается только на тех данных, которые попали в его подвыборку.
   Оно строит разбиения и узлы на основе признаков этих данных, пытаясь "запомнить" зависимости между входными данными (признаками) и целевой переменной

3. Тестирование на OOB-записях
* После обучения дерева его OOB-записи подаются на вход дерева для проверки.
Поскольку дерево не видело эти записи при обучении, оно пытается предсказать их классы, основываясь только на тех правилах, которые оно "выучило" на обучающей подвыборке.

4. Сравнение предсказаний
* Для каждой OOB-записи сравнивается её предсказанный класс с её истинным классом.
- Например:
  - Истинный класс: 1
  - Дерево предсказало: 1
  - Предсказание правильное

## Формула для OOB-метрики

$$
\text{OOB-метрика} = \frac{\text{количество правильных предсказаний}}{\text{общее количество записей}}
$$

```python
oob_score = rf_model.oob_score_
print(f"OOB-оценка модели: {oob_score:.4f}")
```

Модель показала высокую OOB-оценку, что говорит о её способности хорошо обобщать данные.

## Оценка точности модели на обучающей и тестовой выборках:

###**Accuracy (Точность)** — это метрика, которая измеряет долю правильных предсказаний модели относительно общего количества предсказаний

Accuracy показывает, насколько часто модель предсказывает правильный класс для входных данных

### Формула для Accuracy

$$
\text{Accuracy} = \frac{\text{Количество правильных предсказаний}}{\text{Общее количество предсказаний}}
$$


### Принцип работы

1. **Предположим, что у нас есть два класса:**
   - Класс 0: Нормальная транзакция.
   - Класс 1: Мошенническая транзакция.

2. **Результаты модели можно представить в виде матрицы ошибок (Confusion Matrix):**

| Истинные классы \ Предсказанные классы | 0 (Нормальная)    | 1 (Мошенническая) |
|---------------------------------------|-------------------|-------------------|
| **0 (Нормальная)**                    | True Positive (TP) | False Negative (FN) |
| **1 (Мошенническая)**                 | False Positive (FP)| True Negative (TN)  |

- **TP (True Positive):** Правильное предсказание класса 1.
- **TN (True Negative):** Правильное предсказание класса 0.
- **FP (False Positive):** Ошибка, когда 0 предсказано как 1.
- **FN (False Negative):** Ошибка, когда 1 предсказано как 0.

3. **Формула через матрицу ошибок:**

$$
\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
$$


```python
# Оценка точности модели на обучающей выборке
acc_train = rf_model.score(X_train, y_train)  # Рассчитываем accuracy (точность) на обучающих данных

# Оценка точности модели на тестовой выборке
acc_test = rf_model.score(X_test, y_test)  # Рассчитываем accuracy на тестовых данных

# Вывод точности на обучающей выборке
print(f'Accuracy на обучении: {acc_train * 100:.2f} %')

# Вывод точности на тестовой выборке
print(f'Accuracy на тесте: {acc_test * 100:.2f} %')
```

В зависимости от результата сравнения:

1. Precision (Точность):

  Precision — это метрика, которая показывает, какая доля предсказаний положительного класса (например, мошеннические транзакции) оказалась правильной. Она используется для оценки того, насколько "чистыми" являются предсказания модели для положительного класса.
  * Доля предсказаний, которые оказались правильными среди всех предсказанных значений данного класса.
  * Для обоих классов (0 и 1) точность равна 1.00, что означает, что модель не совершила ни одной ошибки среди предсказаний.

###Формула Precision

$$
\text{Precision} = \frac{TP}{TP + FP}
$$
* TP (True Positive): Количество объектов, которые модель правильно предсказала как положительные.
* FP (False Positive): Количество объектов, которые модель ошибочно предсказала как положительные, хотя на самом деле они отрицательные.


2. Recall (Полнота):

  Recall — это метрика, которая показывает, какую долю объектов положительного класса модель правильно нашла. Она используется для оценки того, насколько хорошо модель умеет обнаруживать все истинные положительные объект

  * Доля правильно предсказанных значений данного класса среди всех истинных значений этого класса.
  * Полнота равна 1.00, что указывает на то, что модель правильно классифицировала все примеры обоих классов.

###Формула Recall  
$$
\text{Recall} = \frac{TP}{TP + FN}
$$

3. F1-Score:

  F1-Score — это метрика, которая объединяет Precision (точность) и Recall (полноту) в одно значение. Она используется, чтобы сбалансировать эти две метрики, особенно в случаях, когда важно учитывать как ложные срабатывания (False Positive), так и пропуски (False Negative).

  * Гармоническое среднее между Precision и Recall, отражает баланс между ними.
  F1-Score для обоих классов равен 1.00, что подтверждает идеальную классификацию на обучающей выборке.

### Формула F1-Score

$$
\text{F1-Score} = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
$$


4. Support:
  Support — это метрика, которая показывает количество истинных объектов (или записей) для каждого класса в наборе данных. Она используется для анализа распределения классов в задаче классификации.

  Support — это количество записей для каждого класса в истинных данных
  $y_{\text{true}}$

  * Количество примеров каждого класса в выборке.
Класс "0" представлен 12441 примерами, класс "1" — 12393 примерами.

Модель правильно классифицировала все примеры в обучающей выборке.

6. Macro Avg и Weighted Avg:

  Macro Avg рассчитывает простое среднее метрики (например, Precision, Recall или F1-Score) по всем классам. При этом каждому классу присваивается одинаковый вес, независимо от количества объектов в классе.

### Формула для Macro Avg:

$$
\text{Macro Avg} = \frac{1}{N} \sum_{i=1}^{N} \text{Metric}_i
$$

**Где:**

- *N* : Общее количество классов.
- $\text{Metric}_i$: Значение метрики (Precision, Recall, F1-Score) для \(i\)-го класса.

  
  Weighted Avg учитывает дисбаланс классов, добавляя каждому классу вес, пропорциональный количеству объектов этого класса в данных. Таким образом, классы с большим количеством объектов оказывают большее влияние на итоговую метрику.

  ### Формула для Weighted Avg:

$$
\text{Weighted Avg} = \frac{\sum_{i=1}^{N} \text{Metric}_i \cdot \text{Support}_i}{\sum_{i=1}^{N} \text{Support}_i}
$$

**Где:**
- *N* : Общее количество классов.
- $\text{Metric}_i$: Значение метрики для \(i\)-го класса.
- $\text{Support}_i$: Количество объектов для \(i\)-го класса.



ROC-кривая (характеристическая кривая работы приемника) — это график, который используется для оценки качества бинарной классификации. Она показывает, как изменяется качество модели при различных порогах классификации.

1. **True Positive Rate (TPR):** Доля верно классифицированных объектов положительного класса.

$$
\text{TPR} = \frac{\text{TP}}{\text{TP} + \text{FN}}
$$



2. **False Positive Rate (FPR):** Доля объектов отрицательного класса, которые были классифицированы как положительные.

$$
\text{FPR} = \frac{\text{FP}}{\text{FP} + \text{TN}}
$$

На графике:
- **TPR** по оси \(y\) (вертикальной).
- **FPR** по оси \(x\) (горизонтальной).


AUC (Area Under Curve) — это площадь под ROC-кривой, которая используется как числовой показатель качества модели.

$$
\text{AUC} = \sum_{i=1}^{n-1} \frac{(x_{i+1} - x_i) \cdot (y_i + y_{i+1})}{2}
$$

Где:
- $x_i, x_{i+1}$: Значения False Positive Rate (FPR) на соседних точках кривой.
- $y_i, y_{i+1}$: Значения True Positive Rate (TPR) на соседних точках кривой.
- n\: Общее количество точек на кривой.





```python
from yellowbrick.classifier import ROCAUC  # Импортируем модуль для построения ROC-кривой и оценки AUC

# Создаём объект для построения ROC-кривой
ROC = ROCAUC(rf_model)  # Передаём обученную модель случайного леса

# Обучаем ROC-объект на тренировочной выборке
ROC.fit(X_train, y_train)  # Обучение для вычисления метрик на тренировочных данных

# Оцениваем качество модели на тестовой выборке
ROC.score(X_test, y_test)  # Расчёт ROC-кривой и AUC на тестовых данных

# Настраиваем размер графика
plt.figure(figsize=(6, 4))  # Устанавливаем размер окна для графика

# Отображаем ROC-кривую
ROC.show()  # Строим и показываем ROC-кривую с AUC
```


Высокое качество модели:

Идеальные значения AUC для обоих классов указывают на то, что модель случайного леса отлично различает классы, минимизируя как ложные срабатывания (FPR), так и пропущенные случаи (FN).
Идеальное обобщение:

Результаты на тестовой выборке (AUC = 1.00) подтверждают, что модель хорошо обобщает данные и способна справляться с новыми примерами.

**Гиперпараметры** — это параметры модели машинного обучения, которые задаются перед обучением и не изменяются в процессе обучения. Они влияют на поведение алгоритма и его способность находить решение. Наприме

Подбор гиперпараметров нужен для:

1. Улучшение качества модели:

  * Подбор гиперпараметров позволяет модели лучше адаптироваться к данным и минимизировать ошибку.
2. Предотвращение переобучения:

  * Правильный выбор гиперпараметров (например, глубины деревьев) помогает избежать переобучения.

3. Оптимизация производительности:

  * Хорошо подобранные гиперпараметры снижают вычислительные затраты и время обучения.


###Основные гиперпараметры, влияющие на качество модели случайного леса:

 - n_estimators — число деревьев в "лесу";

 - max_features — число признаков, по которым ищется разбиение;

- min_samples_leaf — минимальное число объектов в листе;

- max_depth — максимальная глубина дерева.


###Метод подбора гиперпараметров

**Grid Search (перебор по сетке)**:
- Как работает:

    * Задаётся диапазон значений для каждого гиперпараметра.
    * Перебираются все возможные комбинации.
    * Для каждой комбинации проводится кросс-валидация, чтобы оценить качество модели.

###Валидационная кривая (Validation Curve)

**Валидационная кривая** — это график, который показывает, как качество модели зависит от значения одного из её гиперпараметров. Она используется для выбора оптимального значения гиперпараметра, чтобы модель была точной, но не переобученной.



##n_estimators

**n_estimators** — это гиперпараметр, который определяет количество базовых моделей (обычно деревьев решений), используемых в ансамблевом методе.

Подбор оптимального значения гиперпараметра n_estimators для модели случайного леса с использованием валидационной кривой. Для ускорения вычислений из обучающего набора данных выбирается случайная подвыборка в 5000 объектов. Далее задаётся диапазон значений n_estimators (100, 300, 500), по которым строится валидационная кривая. Для каждого значения проводится кросс-валидация с двумя фолдами, и метрика точности (accuracy) используется для оценки качества модели.

В процессе кросс-валидации подвыборка данных делится на два фолда: на каждом этапе одна часть данных используется для обучения модели, а другая — для оценки её точности. Таким образом, каждый фолд поочерёдно становится тестовым, что позволяет уменьшить влияние случайности при обучении. Результаты кросс-валидации усредняются для каждого значения n_estimators, и эти значения отображаются на графике. После построения графика код выбирает значение n_estimators, которое обеспечивает наилучший баланс между обучением и обобщением.

```python
from sklearn.ensemble import RandomForestClassifier
from yellowbrick.model_selection import ValidationCurve
import numpy as np
import matplotlib.pyplot as plt

# Уменьшение выборки для ускорения
sample_size = 5000
X_train_sampled = X_train_main.sample(n=sample_size, random_state=21)
y_train_sampled = y_train_main.loc[X_train_sampled.index]

# Настройка параметров n_estimators
n_estimators_grid = [100, 300, 500]
viz_n_estimators = ValidationCurve(
    RandomForestClassifier(random_state=21),
    param_name="n_estimators",
    param_range=n_estimators_grid,
    cv=2,  # Уменьшение количества фолдов
    scoring="accuracy"
)

# Строим кривую
viz_n_estimators.fit(X_train_sampled, y_train_sampled)
viz_n_estimators.show()

# Определяем лучшее значение
n_estimators_test_scores = viz_n_estimators.test_scores_mean_
best_n_estimators = n_estimators_grid[np.argmax(n_estimators_test_scores)]
print(f"Лучшее количество деревьев: {best_n_estimators}")
```

Оптимальное значение: Лучшее значение метрики достигается при n_estimators =  500. Увеличение количества деревьев не дает значительного улучшения, но увеличивает вычислительные затраты.

##max_depth

**max_depth** — это гиперпараметр, который задаёт максимальную глубину дерева решений


Подбор оптимального значения гиперпараметра **max_depth** (максимальной глубины деревьев) в модели случайного леса. Для ускорения вычислений из обучающего набора данных выбирается подвыборка в 5000 объектов. Затем задаётся диапазон значений для max_depth (от 5 до 15), и с помощью валидационной кривой оценивается качество модели для каждого значения глубины. Оценка производится с использованием трёхфолдной кросс-валидации, а метрикой качества выступает точность (accuracy).

Код строит график зависимости точности модели от максимальной глубины деревьев, что позволяет визуально проанализировать, как изменяется качество модели. После этого определяется оптимальное значение max_depth, которое соответствует наибольшей точности на валидационной выборке. Такой подход помогает найти баланс между переобучением (слишком большая глубина) и недообучением (слишком малая глубина).

```python
from sklearn.ensemble import RandomForestClassifier
from yellowbrick.model_selection import ValidationCurve
import numpy as np
import matplotlib.pyplot as plt

# Уменьшаем выборку для ускорения вычислений
sample_size = 5000  # Подвыборка данных
X_train_sampled = X_train_main.sample(n=sample_size, random_state=21)
y_train_sampled = y_train_main.loc[X_train_sampled.index]

# Диапазон значений для настройки max_depth
max_depth_grid = np.arange(5, 15)

# Создаём визуализацию валидационной кривой для max_depth
viz_max_depth = ValidationCurve(
    RandomForestClassifier(n_estimators=500, random_state=21),  # Используем лучшее значение n_estimators
    param_name="max_depth",                                   # Настраиваемый параметр - max_depth
    param_range=max_depth_grid,                               # Диапазон значений
    cv=3,                                                    # Кросс-валидация с 3 фолдами
    scoring="accuracy"                                       # Метрика оценки - точность
)

# Строим кривую
viz_max_depth.fit(X_train_sampled, y_train_sampled)
viz_max_depth.show()

# Определяем лучшее значение max_depth
max_depth_test_scores = viz_max_depth.test_scores_mean_
best_max_depth = max_depth_grid[np.argmax(max_depth_test_scores)]
print(f"Лучшая глубина деревьев: {best_max_depth}")
```

Оптимальная глубина деревьев (max_depth): 14. На этом значении достигается наилучший баланс между точностью на обучающей и тестовой выборках.

##min_samples_leaf

**min_samples_leaf** — это гиперпараметр, который задаёт минимальное количество объектов (примеров) в каждом листе дерева решений. Подбирает оптимальное значение гиперпараметра min_samples_leaf для модели случайного леса. Этот параметр задаёт минимальное количество объектов, которые должны быть в каждом листе дерева, чтобы разбиение считалось допустимым. Для ускорения вычислений из обучающего набора данных выбирается подвыборка из 5000 записей. Далее задаётся диапазон значений для min_samples_leaf (от 1 до 15), и с помощью валидационной кривой проводится оценка точности модели на каждом значении этого гиперпараметра.

Кросс-валидация с 3 фолдами используется для проверки точности на каждом этапе, а метрика точности (accuracy) помогает выбрать оптимальное значение параметра. После построения кривой зависимости точности от значений min_samples_leaf код находит и выводит значение, при котором достигается максимальная точность модели. Это помогает найти баланс между сложностью модели и её способностью обобщать данные.

##MDI

MDI (Mean Decrease in Impurity) — это метод оценки важности признаков в модели случайного леса (Random Forest). Он основан на уменьшении неопределённости (или нечистоты) в данных, которое вносит каждый признак на этапе построения дерева решений. Чем больше признак уменьшает нечистоту узлов, тем более значимым он считается для модели.



### **1. Измерение нечистоты узлов**

#### **Индекс Джини:**
Формула для узла $s$ с $K$ классами:
$$
G(s) = 1 - \sum_{k=1}^{K} p_k^2
$$
где:
- $p_k$ — доля объектов класса $k$ в узле $s$,
- $K$ — общее количество классов.

#### **Энтропия:**
Формула для узла $s$:
$$
H(s) = - \sum_{k=1}^{K} p_k \log_2(p_k)
$$
где:
- $p_k$ — доля объектов класса $k$,
- $K$ — общее количество классов.


### **2. Уменьшение нечистоты (Decrease in Impurity)**

Для узла $s$, где используется признак $X_j$, уменьшение нечистоты вычисляется как:
$$
\Delta \text{Impurity}(s) = \text{Impurity}(s_{\text{parent}}) - \Big( \frac{N_{\text{left}}}{N_{\text{parent}}} \cdot \text{Impurity}(s_{\text{left}}) + \frac{N_{\text{right}}}{N_{\text{parent}}} \cdot \text{Impurity}(s_{\text{right}}) \Big)
$$
где:
- $\text{Impurity}(s_{\text{parent}})$ — нечистота родительского узла,
- $\text{Impurity}(s_{\text{left}})$ и $\text{Impurity}(s_{\text{right}})$ — нечистоты дочерних узлов,
- $N_{\text{parent}}, N_{\text{left}}, N_{\text{right}}$ — количество объектов в родительском и дочерних узлах.


### **3. Расчёт MDI для признака $X_j$**

MDI для признака $X_j$ вычисляется как сумма уменьшений нечистоты по всем узлам дерева, где использовался этот признак, усреднённая по количеству деревьев:
$$
\text{MDI}(X_j) = \frac{1}{T} \sum_{t=1}^{T} \sum_{s \in S(X_j)} \Delta \text{Impurity}(s)
$$
где:
- $T$ — количество деревьев в случайном лесу,
- $S(X_j)$ — множество узлов, где использовался признак $X_j$,
- $\Delta \text{Impurity}(s)$ — уменьшение нечистоты в узле $s$.

##Перемешивание признаков
Перемешивание признаков (Feature Shuffling) — это метод оценки важности признаков в модели машинного обучения, который основан на случайной перестановке значений одного признака в данных. Цель этого метода — определить, насколько сильно модель зависит от конкретного признака. Если после перемешивания важного признака точность модели резко снижается, то это указывает на высокую значимость признака.

### Как происходит перемешивние

### Изначальный набор данных

| $X_1$ | $X_2$ | $X_3$ | $y$ |
|--------|--------|--------|------|
| 10     | 5      | 1      | 0    |
| 15     | 7      | 0      | 1    |
| 12     | 6      | 1      | 0    |
| 20     | 8      | 1      | 1    |

Здесь:
- $X_1, X_2, X_3$ — признаки,
- $y$ — целевая переменная.



### Перемешивание признака $X_2$

Мы случайным образом перемешиваем значения признака $X_2$, сохраняя остальные признаки и целевую переменную без изменений.

| $X_1$ | $X_2$ | $X_3$ | $y$ |
|--------|--------|--------|------|
| 10     | 7      | 1      | 0    |
| 15     | 8      | 0      | 1    |
| 12     | 5      | 1      | 0    |
| 20     | 6      | 1      | 1    |

Теперь связь между $X_2$ и целевой переменной $y$ разрушена, так как значения $X_2$ переставлены случайным образом.


**permutation_importance** — это метод оценки важности признаков, который помогает понять, насколько сильно модель зависит от каждого признака. Принцип работы основан на случайном перемешивании значений одного признака в тестовом наборе данных. После этого модель повторно делает предсказания, и рассчитывается новая метрика качества (например, точность, F1-Score или ROC-AUC). Разница между базовой метрикой (до перемешивания) и метрикой после перемешивания показывает, насколько важен этот признак. Чем больше снижение качества модели, тем более значим признак.

Метод повторяет процесс несколько раз для повышения точности оценки (количество повторений задаётся параметром n_repeats). В результате возвращается среднее значение изменения метрики (importances_mean) и стандартное отклонение (importances_std) для каждого признака. Этот подход независим от типа модели и предоставляет интерпретируемые результаты. Однако он может быть вычислительно затратным, особенно на больших наборах данных, и чувствителен к взаимозависимости признаков.


```python
from sklearn.inspection import permutation_importance

# Вычисляем важность признаков методом перестановки для модели `rf_clf_opt` на данных X_test_main и y_test_main.
# n_repeats=5 — число повторений перемешивания для каждого признака.
# random_state=21 — фиксируем случайное состояние для воспроизводимости.
# n_jobs=-1 — используем все доступные процессоры для ускорения расчётов.
permutation_importance_result = permutation_importance(
    rf_clf_opt, X_test_main, y_test_main,
    n_repeats=5,
    random_state=21,
    n_jobs=-1
)

# Просмотр результата permutation importance
# `permutation_importance_result` содержит три атрибута:
# - importances: массив уменьшений метрики для каждого признака и повторения.
# - importances_mean: средние значения уменьшения метрики для каждого признака.
permutation_importance_result

# Сортировка признаков по средней важности
# Отсортированный по возрастанию массив индексов признаков на основе `importances_mean`.
sorted_id = np.argsort(permutation_importance_result.importances_mean)

# Средние значения важности признаков, отсортированные по возрастанию.
# Показывает, насколько каждый признак уменьшает метрику модели в среднем.
permutation_importance_result.importances_mean[sorted_id]

# Вывод имен признаков в порядке убывания их важности
# Список имён признаков, отсортированных по убыванию значимости.
features[sorted_id]
```

* Последние элементы (V4 и V14) имеют наибольшую значимость. Они оказывают существенное влияние на предсказания, и их удаление существенно снизит качество модели.

#**Раздел 4. Оценивание качества алгоритма машинного обучения**

##**4.1 Обоснование выбора метрик для оценивания качества обученной модели машинного обучения**


#### **1. Accuracy (Точность классификации)**
- **Что измеряет**: Это базовая метрика, которая вычисляет долю правильно классифицированных объектов от общего числа объектов в выборке.
- **Обоснование**: Accuracy предоставляет общее представление о том, насколько хорошо модель справляется с классификацией. Однако эта метрика может быть обманчивой в задачах с несбалансированными классами, где большая доля объектов относится к одному классу. Например, если большинство транзакций нормальные, высокая точность может не отражать способность модели корректно выявлять мошенничество.



#### **2. Precision (Точность)**
- **Что измеряет**: Precision показывает, какую долю из всех объектов, предсказанных моделью как положительные (например, мошеннические транзакции), составляют действительно положительные объекты.
- **Обоснование**: Precision важен, когда критично минимизировать ложные срабатывания (False Positives). В контексте мошенничества это уменьшает количество нормальных транзакций, ошибочно классифицированных как мошеннические, что снижает неудобства для клиентов.



#### **3. Recall (Полнота)**
- **Что измеряет**: Recall вычисляет, какую долю из всех истинных объектов положительного класса модель смогла правильно классифицировать.
- **Обоснование**: Полнота играет ключевую роль в задачах, где важно выявить как можно больше объектов положительного класса (например, всех мошеннических транзакций). Пропуск мошеннической активности (False Negatives) может привести к финансовым потерям, поэтому Recall особенно важен в подобных задачах.



#### **4. F1-Score**
- **Что измеряет**: F1-Score — это гармоническое среднее между Precision и Recall. Оно обеспечивает баланс между этими двумя метриками.
- **Обоснование**: Эта метрика особенно полезна, если классы несбалансированы. F1-Score помогает найти оптимальное соотношение между точностью и полнотой, чтобы одновременно минимизировать как ложные срабатывания, так и пропуски мошеннических транзакций.



#### **5. ROC-AUC (Площадь под ROC-кривой)**
- **Что измеряет**: Эта метрика показывает способность модели различать классы, независимо от выбранного порога классификации. ROC-AUC строится на основе истинно положительных и ложно положительных долей (TPR и FPR).
- **Обоснование**: ROC-AUC позволяет оценить общую производительность модели. Она полезна в задачах, где необходимо понять, насколько хорошо модель справляется с различением классов, вне зависимости от порога. Это особенно важно для задач с несбалансированными данными.



#### **6. OOB (Out-of-Bag) оценка**
- **Что измеряет**: OOB оценка — это встроенный механизм случайного леса, который использует данные, не вошедшие в подвыборку для обучения конкретного дерева, для проверки его качества.
- **Обоснование**: Это удобный способ оценки модели без необходимости выделения тестового набора. OOB позволяет оценивать производительность модели, экономя данные, что особенно полезно при ограниченном размере выборки.



#### **7. Permutation Importance (Перемешивание признаков)**
- **Что измеряет**: Этот метод оценивает важность каждого признака, переставляя его значения случайным образом и оценивая влияние на производительность модели.
- **Обоснование**: Permutation Importance полезен для анализа интерпретируемости модели. Он помогает определить, какие признаки вносят наибольший вклад в качество модели. Это особенно важно в задачах, где необходимо объяснять работу модели (например, в финансовой сфере).



#### **8. Macro Avg**
- **Что измеряет**: Это среднее значение выбранной метрики (Precision, Recall, F1-Score) по всем классам без учёта их пропорций.
- **Обоснование**: Macro Avg используется, когда необходимо равномерно оценить производительность модели по всем классам, даже если они представлены в данных неравномерно.



#### **9. Weighted Avg**
- **Что измеряет**: Это среднее значение метрик с учётом пропорций объектов в каждом классе.
- **Обоснование**: Weighted Avg особенно полезна в задачах с несбалансированными классами, так как она учитывает дисбаланс классов и предоставляет более точную оценку производительности модели в реальных условиях.

### **10.  MDI (Mean Decrease Impurity)**
-  **Что измеряет**
  * MDI оценивает важность признаков, основываясь на уменьшении неопределённости (impurity) при разбиении данных в деревьях. Для каждого признака вычисляется вклад в уменьшение неоднородности классов на каждом шаге разбиения, а итоговая оценка усредняется по всем деревьям случайного леса.

- **Обоснование**
  - **Интерпретируемость:** MDI позволяет понять, насколько каждый признак влияет на улучшение модели, показывая его вклад в уменьшение неопределённости.
  - **Простота:** MDI автоматически вычисляется во время обучения случайного леса, что делает его удобным для анализа.
  - **Широкая применимость:** Подходит для различных задач классификации и регрессии, особенно с большим числом признаков.


##**4.2 Описание математических выражений выбранных метрик качества модели машинного обучения**

### 1. Accuracy

* **Accuracy** показывает, насколько часто модель предсказывает правильный класс для входных данных

$$
\text{Accuracy} = \frac{\text{Количество правильных предсказаний}}{\text{Общее количество предсказаний}}
$$





### 2. Точность (Precision)

  - **Определение:** Доля правильных мошеннических транзакций среди всех предсказанных как мошеннические.

  $$
  Precision = \frac{TP}{TP + FP}
  $$

  - **Где:**
    - **TP (True Positive):** Истинно положительные предсказания (правильные предсказания мошеннических транзакций).
    - **FP (False Positive):** Ложные срабатывания (нормальные транзакции, ошибочно помеченные как мошеннические).



### 3. Полнота (Recall)

  - **Определение:** Доля правильно предсказанных мошеннических транзакций среди всех истинных мошеннических транзакций.

  $$
  Recall = \frac{TP}{TP + FN}
  $$

  - **Где:**
    - **FN (False Negative):** Пропущенные мошеннические транзакции (ошибочно предсказанные как нормальные).



### 4. F1-мера

  - **Определение:** Гармоническое среднее точности и полноты.

  $$
  F1 = 2 \cdot \frac{Precision \cdot Recall}{Precision + Recall}
  $$


### 5. ROC-AUC (Receiver Operating Characteristic - Area Under Curve)

- AUC (Area Under Curve) — это площадь под ROC-кривой, которая используется как числовой показатель качества модели.

$$
\text{AUC} = \sum_{i=1}^{n-1} \frac{(x_{i+1} - x_i) \cdot (y_i + y_{i+1})}{2}
$$

Где:
- $x_i, x_{i+1}$: Значения False Positive Rate (FPR) на соседних точках кривой.
- $y_i, y_{i+1}$: Значения True Positive Rate (TPR) на соседних точках кривой.
- n\: Общее количество точек на кривой

### 6. OOB-оценка (Out-of-Bag Error)

  - **Определение:** Оценка качества модели, вычисляемая на данных, которые не участвовали в обучении каждого отдельного дерева в случайном лесу (Out-of-Bag data).


$$
\text{OOB-метрика} = \frac{\text{количество правильных предсказаний}}{\text{общее количество записей}}
$$

  

### 7. Macro Avg
Macro Avg рассчитывает простое среднее метрики (например, Precision, Recall или F1-Score) по всем классам. При этом каждому классу присваивается одинаковый вес, независимо от количества объектов в классе.


$$
\text{Macro Avg} = \frac{1}{N} \sum_{i=1}^{N} \text{Metric}_i
$$

**Где:**

- *N* : Общее количество классов.
- $\text{Metric}_i$: Значение метрики (Precision, Recall, F1-Score) для \(i\)-го класса.

### 8. Weighted Avg

$$
\text{Weighted Avg} = \frac{\sum_{i=1}^{N} \text{Metric}_i \cdot \text{Support}_i}{\sum_{i=1}^{N} \text{Support}_i}
$$

**Где:**
- *N* : Общее количество классов.
- $\text{Metric}_i$: Значение метрики для \(i\)-го класса.
- $\text{Support}_i$: Количество объектов для \(i\)-го класса.

### 9. Permutation Importance

Если  $M$ — метрика качества модели (например, Accuracy), а $X_j$ — значения признака $j$, то:

$$
\Delta M_j = M_{\text{baseline}} - M_{\text{shuffled}}(X_j)
$$

где:
-  $M_{\text{baseline}}$ — качество модели на исходных данных.
-  $M_{\text{shuffled}}(X_j)$ — качество модели после перемешивания значений признака $X_j$.
- $\Delta M_j$ — снижение качества модели, которое интерпретируется как важность признака $X_j$.

### 10. Формула для MDI (Mean Decrease Impurity)

$$
\text{MDI}(X_j) = \frac{1}{T} \sum_{t=1}^{T} \sum_{n \in N_j^t} \Delta I_n^t
$$

где:
- $ T $ — общее количество деревьев в случайном лесу.
- $ N_j^t $ — множество узлов дерева $t$, где использовался признак $X_j$ для разбиения.
- $\Delta I_n^t$ — уменьшение impurity (например, индекса Джини или энтропии) в узле $n$ дерева $t$:

$$
\Delta I_n^t = I_{\text{parent}} - \left( w_{\text{left}} \cdot I_{\text{left}} + w_{\text{right}} \cdot I_{\text{right}} \right)
$$

где:
- $ I_{\text{parent}}, I_{\text{left}}, I_{\text{right}} $ — $impurity$ для родительского узла, левого и правого дочерних узлов соответственно.
- $w_{\text{left}}, w_{\text{right}} $ — веса узлов (доли объектов в каждом дочернем узле от общего числа объектов в родительском узле).

Это указывает на минимальное количество ошибок при высокой эффективности модели.

Отчёт классификации подкрепляет эти выводы. На обучающей выборке метрики Precision, Recall и F1-Score для каждого класса достигают значения 0.99 или выше, а общая точность составляет 99.9%. На тестовой выборке результаты также остаются стабильными: метрики Precision, Recall и F1-Score для обоих классов — 0.99, общая точность — 99.8%. Это говорит о том, что модель способна хорошо различать классы даже на новых данных, сохраняя баланс между точностью предсказаний и полнотой обнаружения.

Этот результат также подтверждается кросс-валидацией, где средняя точность по пяти фолдам составила 0.9989, демонстрируя стабильность и устойчивость модели при работе с различными подвыборками данных. Кроме того, OOB-метрика, равная 0.9983, подтверждает, что модель хорошо обучается на предоставленных данных без необходимости дополнительного тестового набора.

Таким образом, модель случайного леса доказала свою эффективность и надёжность для классификации транзакций, минимизируя ошибки и показывая стабильные результаты на различных метриках, включая Precision, Recall, F1-Score, ROC-AUC и кросс-валидацию.
