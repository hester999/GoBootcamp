```python
import matplotlib.pyplot as plt

# Построение боксплотов для всех признаков на одном полотне и удаление выбросов
plt.figure(figsize=(10, 50))  # Устанавливаем размер графика для визуализации

# Проходим по каждому признаку из списка числовых признаков
for i, column in enumerate(numeric_columns, 1):  # enumerate даёт индекс и название признака
    # Расчет первого (Q1) и третьего (Q3) квартилей
    Q1 = df[column].quantile(0.25)  # Первый квартиль (25-й процентиль)
    Q3 = df[column].quantile(0.75)  # Третий квартиль (75-й процентиль)
    IQR = Q3 - Q1  # Интерквартильный размах (разница между Q3 и Q1)

    # Определение границ выбросов
    lower_bound = Q1 - 1.5 * IQR  # Нижняя граница (в пределах 1.5 интерквартильных размахов от Q1)
    upper_bound = Q3 + 1.5 * IQR  # Верхняя граница (в пределах 1.5 интерквартильных размахов от Q3)

    # Подсчёт количества выбросов в текущем признаке
    outliers = ((df[column] < lower_bound) | (df[column] > upper_bound)).sum()

    # Удаление строк с выбросами для текущего признака
    df = df[~((df[column] < lower_bound) | (df[column] > upper_bound))]

    # Построение боксплота для текущего признака
    plt.subplot(len(numeric_columns), 1, i)  # Создаем отдельный график для каждого признака
    plt.boxplot(df[column], vert=False, patch_artist=True)  # Рисуем горизонтальный боксплот
    plt.title(f'{column} (Outliers: {outliers})')  # Указываем имя признака и количество выбросов
    plt.xlabel('Value')  # Метка оси X
    plt.grid(axis='x')  # Добавляем сетку для удобства чтения графика

# Уплотнение макета графиков
plt.tight_layout()

# Отображение всех графиков
plt.show()
```

Квартиль — это статистический показатель, который делит упорядоченный набор данных на четыре равные части. Каждая часть содержит примерно 25% данных.

Первый квартиль (Q1):

* Значение, ниже которого находится 25% всех данных.
* Это 25-й процентиль.

Медиана (Q2):

* Это второй квартиль, делящий данные пополам.
* Значение, ниже которого находится 50% всех данных.
* Медиана также известна как 50-й процентиль.

Третий квартиль (Q3):

* Значение, ниже которого находится 75% всех данных.
* Это 75-й процентиль.

```python
import pandas as pd  # Импорт библиотеки для работы с данными

def drop_features_by_correlation(df, target_column, min_corr=0.1, max_corr=0.9):
    """
    Удаляет признаки из датафрейма, если их корреляция с целевой переменной меньше min_corr
    или больше max_corr.

    :param df: pd.DataFrame, входной датафрейм
    :param target_column: str, название целевого столбца
    :param min_corr: float, минимальное значение корреляции (по модулю)
    :param max_corr: float, максимальное значение корреляции (по модулю)
    :return: pd.DataFrame, датафрейм с удалёнными признаками
    """

    # Вычисляем корреляционную матрицу для всех признаков
    correlation_matrix = df.corr()

    # Извлекаем корреляцию признаков с целевой переменной
    target_correlation = correlation_matrix[target_column].abs()

    # Отбираем признаки, которые имеют корреляцию меньше min_corr или больше max_corr
    to_drop = target_correlation[
        (target_correlation < min_corr) | (target_correlation > max_corr)
    ].index  # Получаем имена таких признаков

    to_drop = to_drop.drop(target_column)  # Исключаем целевой столбец из списка для удаления

    # Печатаем список признаков, которые будут удалены
    print(f"Признаки для удаления: {list(to_drop)}")

    # Удаляем признаки с неподходящей корреляцией
    df = df.drop(columns=to_drop)

    # Возвращаем очищенный датафрейм
    return df
```

Отбираются признаки, корреляция которых:
* Меньше минимального значения (min_corr) — они мало связаны с целевым классом и могут не быть полезными для модели.
* Больше максимального значения (max_corr) — слишком высокая корреляция может привести к дублированию информации.


Зачем это нужно:
* Удаление нерелевантных признаков:

  1. Признаки с низкой корреляцией (например, <0.1) не содержат значимой информации для предсказания целевого класса. Их удаление улучшает производительность модели.

* Устранение мультиколлинеарности:

  1. Признаки с высокой корреляцией между собой дублируют информацию. Удаление
таких признаков снижает избыточность данных.

* Упрощение модели:

  1. Удаление ненужных признаков уменьшает размерность данных, что снижает риск переобучения.

# Обучение модели с гиперпараметрами по умолчанию

```python
from sklearn.model_selection import train_test_split  # Импорт функции для разделения данных на обучающую и тестовую выборки
from sklearn.ensemble import RandomForestClassifier  # Импорт классификатора случайного леса
from sklearn.metrics import classification_report, confusion_matrix  # Метрики для оценки качества модели
import numpy as np  # Импорт библиотеки numpy для работы с массивами (здесь пока не используется)

# Разделяем признаки (X) и целевую переменную (y)
X = df.drop(columns=['Class'])  # Удаляем столбец 'Class', оставляя только признаки
y = df['Class']  # Целевая переменная (мошеннические/нормальные транзакции)

# Разделяем данные на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(
    X, y,  # Входные данные (X) и целевая переменная (y)
    test_size=0.3,  # Размер тестовой выборки — 30% от общего объёма данных
    random_state=21  # Фиксируем случайное состояние для воспроизводимости
)

# Создаём модель случайного леса
rf_model = RandomForestClassifier(
    oob_score=True,  # Используем "Out-of-Bag" оценку для проверки качества модели
    n_jobs=-1,  # Используем все доступные процессоры для ускорения обучения
    random_state=21  # Фиксируем случайное состояние для воспроизводимости
)
X_train_main = X_train.copy()
X_test_main =  X_test.copy()
y_train_main = y_train.copy()
y_test_main = y_test.copy()

# Обучаем модель на обучающих данных
rf_model.fit(X_train, y_train)  # Передаём обучающие признаки и целевую переменную
```

##Кросс-валидация

```python
from sklearn.model_selection import cross_val_score  # Импорт функции для проведения кросс-валидации

# Используем 5-кратную кросс-валидацию
cv_scores = cross_val_score(  # Функция возвращает оценки качества модели для каждой из 5 частей
    rf_model,  # Модель случайного леса, которую мы хотим проверить
    X,         # Все признаки из датасета
    y,         # Целевая переменная
    cv=5       # Число разбиений данных (5-кратная кросс-валидация)
)

# Выводим оценки для каждой итерации кросс-валидации
print(f"Кросс-валидация: {cv_scores}")

# Вычисляем среднюю оценку по всем итерациям кросс-валидации
print(f"Средняя оценка кросс-валидации: {cv_scores.mean()}")
```

Кросс-валидация — это метод оценки качества модели, который проверяет её устойчивость и способность обобщать данные. Она позволяет:

1. Проверить стабильность модели: Оценивается, насколько результаты на разных разбиениях данных близки друг к другу. Низкая дисперсия указывает на стабильность.

2. Избежать переобучения: Тестирование на новых подвыборках помогает убедиться, что модель не "запоминает" обучающие данные, а способна предсказывать на новых данных.

3. Получить более точную оценку: Среднее значение кросс-валидации предоставляет надёжное представление о производительности модели, в отличие от однократного тестирования.

4. Обеспечить равномерное распределение данных: Разбиение данных на части позволяет учесть их дисбаланс или особенности, повышая надёжность оценки.

5. Подтвердить качество модели: Сравнение метрик кросс-валидации и тестовой выборки подтверждает корректное обучение модели и надёжность результатов.

1. Модель стабильна: Результаты кросс-валидации демонстрируют, что модель уверенно предсказывает вне зависимости от разбиения данных.
2. Высокое качество: Средняя точность 99.89% говорит о том, что модель эффективно справляется с задачей классификации.
3. Отсутствие переобучения (предположение): Стабильные результаты на кросс-валидации позволяют предположить, что модель не переобучается.

# Линейные модели


```python
# Получение списка всех столбцов (признаков) из датафрейма df1
features = df1.columns.tolist()  # Метод .columns возвращает имена столбцов, .tolist() преобразует их в список

# Вывод списка признаков (названий столбцов) для проверки
features  # Отображаем результат в виде списка
```

```python
# Список числовых признаков (features) из датафрейма
num_cols = [  # Создаём список имен столбцов, которые являются числовыми признаками
    'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10', 'V11',  # Первые 11 признаков
    'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20', 'V21',  # Следующие 10 признаков
    'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount'  # Остальные числовые признаки
]

# Назначаем числовые признаки как входные данные для модели
feature_cols = num_cols  # Устанавливаем список числовых признаков для дальнейшего использования в модели

# Указываем целевую переменную (target), по которой модель будет предсказывать
target_col = 'Class'  # 'Class' - это целевая переменная (например, 0 - нормальные, 1 - мошеннические транзакции)
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV  # Импорт моделей логистической регрессии
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold  # Импорт инструментов для разделения данных и поиска гиперпараметров
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder  # Импорт инструментов для предварительной обработки данных
from sklearn.pipeline import make_pipeline  # Импорт функции для создания пайплайнов
from sklearn.compose import make_column_transformer  # Импорт функции для обработки столбцов с различными трансформациями

# Создание трансформера для числовых данных
column_transformer = make_column_transformer(
    (RobustScaler(), num_cols)  # Применяем RobustScaler для нормализации числовых признаков
)

# Создание пайплайна, который включает трансформацию столбцов и модель логистической регрессии
pipeline = make_pipeline(column_transformer, LogisticRegression())  # Пайплайн: трансформация -> модель

# Настройка кросс-валидации с использованием стратифицированного разбиения
kfolds = StratifiedKFold(n_splits=5, shuffle=True, random_state=10)  # 5 фолдов с перемешиванием данных

# Сетка параметров для поиска оптимального значения параметра регуляризации (C)
param_grid = {
    "logisticregression__C": [100, 10, 1, 0.1, 0.01, 0.001]  # Значения параметра C, где меньшее значение -> сильнее регуляризация
}

# Настройка поиска гиперпараметров с использованием GridSearchCV
gs = GridSearchCV(
    pipeline,  # Пайплайн для оптимизации
    param_grid,  # Сетка параметров
    scoring="roc_auc",  # Используем ROC-AUC как метрику для оценки
    cv=kfolds,  # Стратифицированная кросс-валидация
    n_jobs=4  # Используем 4 потока для ускорения вычислений
)
```

```python
# Обучение модели с использованием GridSearchCV
gs = gs.fit(df1[feature_cols], df1[target_col])  # Передаём признаки и целевую переменную для обучения
```

```python
# Получение наилучшего результата (лучшего ROC-AUC) после кросс-валидации
gs.best_score_  # Выводим максимальное значение метрики ROC-AUC, достигнутое моделью
```

```python

# Получение наилучших гиперпараметров, найденных GridSearchCV
gs.best_params_  # Выводим параметры модели, которые обеспечили лучший результат
```

Модель логистической регрессии показала отличный результат (ROC-AUC = 99.34%), что свидетельствует о её высокой эффективности в задаче классификации.

Наилучшее значение параметра регуляризации C = 100 подтверждает, что модель лучше работает с минимальной регуляризацией, не ограничивая сильно веса признаков.

```python
# Импорт библиотеки CatBoost для работы с градиентным бустингом
import catboost

# Разделение данных на тренировочную и валидационную выборки
splitted_data = train_test_split(
    df1[feature_cols],  # Признаки для обучения
    df1[target_col],    # Целевая переменная
    test_size=0.25,     # 25% данных оставляем для валидации
    stratify=df1[target_col],  # Сохраняем пропорции классов в выборках
    random_state=10     # Фиксируем случайное состояние для воспроизводимости
)

# Распаковываем разделённые данные в соответствующие переменные
X_train, X_val, y_train, y_val = splitted_data  # Признаки и метки для обучения и валидации

# Инициализация классификатора CatBoost с указанием метрики качества
cat_classifier = catboost.CatBoostClassifier(eval_metric="AUC")  # Используем метрику AUC для оценки

# Создание объектов Pool для CatBoost из тренировочных данных
train_dataset = catboost.Pool(data=X_train, label=y_train)  # Пул данных для обучения

# Создание объектов Pool для CatBoost из валидационных данных
eval_dataset = catboost.Pool(data=X_val, label=y_val)  # Пул данных для валидации
```

```python
# Обучение модели CatBoostClassifier
model = cat_classifier.fit(
    train_dataset,  # Данные для обучения (тренировочный пул)
    eval_set=eval_dataset  # Данные для валидации (валидационный пул)
)

# Вывод модели после обучения
model  # Показывает конфигурацию обученной модели и метрики
```

##min_samples_leaf

```python
from sklearn.ensemble import RandomForestClassifier
from yellowbrick.model_selection import ValidationCurve
import numpy as np
import matplotlib.pyplot as plt

# Уменьшаем выборку для ускорения вычислений
sample_size = 5000  # Подвыборка данных
X_train_sampled = X_train_main.sample(n=sample_size, random_state=21)
y_train_sampled = y_train_main.loc[X_train_sampled.index]

# Диапазон значений для настройки min_samples_leaf
min_samples_leaf_grid = np.arange(1, 15)

# Создаём визуализацию валидационной кривой для min_samples_leaf
viz_min_samples_leaf = ValidationCurve(
    RandomForestClassifier(n_estimators=500, max_depth=14, random_state=21),  # Используем лучшие n_estimators и max_depth
    param_name="min_samples_leaf",                                           # Настраиваемый параметр
    param_range=min_samples_leaf_grid,                                       # Диапазон значений
    cv=3,                                                                   # Кросс-валидация с 3 фолдами
    scoring="accuracy"                                                      # Метрика оценки - точность
)

# Строим кривую
viz_min_samples_leaf.fit(X_train_sampled, y_train_sampled)
viz_min_samples_leaf.show()

# Определяем лучшее значение min_samples_leaf
min_samples_leaf_test_scores = viz_min_samples_leaf.test_scores_mean_
best_min_samples_leaf = min_samples_leaf_grid[np.argmax(min_samples_leaf_test_scores)]
print(f"Лучшее минимальное количество объектов в листе: {best_min_samples_leaf}")
```

При значении min_samples_leaf = 1 точность на обучении достигает максимума (1.0), что указывает на то, что модель способна идеально запомнить обучающие данные. С увеличением значения min_samples_leaf точность на обучении снижается. Это объясняется тем, что минимальное количество объектов в листе заставляет модель упрощать деревья и уменьшать их глубину, что снижает риск переобучения.

##MDI

Mean Decrease in Impurity (MDI) — это метрика, используемая в ансамблевых методах (например, Random Forest), для измерения важности каждого признака. Она вычисляется как сумма уменьшения неопределенности (impurity), вызванного использованием конкретного признака в разбиениях дерева

```python
mdi_importance = rf_clf_opt.feature_importances_  # Получение важности признаков из обученной модели Random Forest
mdi_importance  # Отображение важностей
```

массив mdi_importance содержит числовые значения, которые показывают вклад каждого признака в классификацию. Чем выше значение, тем больше влияние.


```python
sorted_id = np.argsort(mdi_importance)  # Получение индексов признаков в порядке возрастания их важности
mdi_importance[sorted_id]  # Отображение важностей признаков в отсортированном порядке
```

Упорядочивает признаки по степени их важности.
Самые низкие значения показывают наименее значимые признаки, а высокие — наиболее важные.
